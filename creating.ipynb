```python
import glob
import folium
from folium.plugins import HeatMap
import pandas as pd

directory_path = "/Users/amagupt29/Documents/Amaira-ds/PedestrianData/"
ls = glob.glob(directory_path + "*.[cC][sS][vV]")

def combine():
    dataframes = []
    for file in ls:
        df = pd.read_csv(file)
        dataframes.append(df)
    return pd.concat(dataframes)

def make_coordinates(df):
    coordinates = []
    for lat, lon in zip(df["LATITUDE"], df["LONGITUD"]):
        if pd.isna(lat) or pd.isna(lon):
            coordinates.append(pd.NA)
        else:
            coordinates.append((lat, lon))
    df["Coordinates"] = coordinates
    return df

def make_dates(df):
    datetimes = []
    for i in range(len(df)):
        try:
            year = int(df.loc[i, "YEAR"])
            month = int(df.loc[i, "MONTH"])
            day = int(df.loc[i, "DAY"])
            hour = int(df.loc[i, "HOUR"])
            minute = int(df.loc[i, "MINUTE"])

            dt = pd.to_datetime(
                f"{year}-{month:02d}-{day:02d} {hour:02d}:{minute:02d}",
                format="%Y-%m-%d %H:%M"
            )
            datetimes.append(dt)
        except Exception:
            datetimes.append(pd.NaT)
    df["UpDATES"] = datetimes
    return df


def create_heatmap(df, map_location, start):
    m = folium.Map(location=map_location, zoom_start=start, min_zoom=4)
    heat_data = []
    for coords in df["Coordinates"]:
        if coords is pd.NA or coords is None:
            continue
        lat, lon = coords
        heat_data.append((lat, lon))
    HeatMap(heat_data, radius=6, blur=1).add_to(m)
    return m
    
combine_result = combine()
combine_result = make_coordinates(combine_result)
combine_result = make_dates(combine_result)
heatmap = create_heatmap(combine_result, [36.77, -119.42], 6)
heatmap  
```
